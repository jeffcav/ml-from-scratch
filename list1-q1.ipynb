```python
from ml.stats import mse
from ml.dataset import load_csv
from ml.models.linear import LinearRegression
from ml.algorithms.optimization import GradientDescent, OrdinaryLeastSquares, StochasticGradientDescent

import matplotlib.pyplot as plt
```

## Q1

```python
X, y = load_csv('./datasets/artificial1d.csv')
```

### Q1.a
#### Train a Linear Regression model using the Ordinary Least Squares

```python
ols = OrdinaryLeastSquares()
linear_regression_ols = LinearRegression(ols)
linear_regression_ols.fit(X, y);
```

```python
y_ols = linear_regression_ols.predict(X)

print("Weights:", linear_regression_ols.get_weights())
print("MSE:", mse(y, y_ols) )

plt.plot(X, y, label='observations')
plt.plot(X, y_ols, label='predictions')
plt.legend()
plt.title('Linear Regression with Ordinary Least Squares')
plt.show()
```

### Q1.b
#### Train a Linear Regression model using the Gradient Descent

```python
epochs = 2000
learning_rate = 0.05
regularization = 0

gd = GradientDescent(epochs, learning_rate, regularization)
linear_regression_gd = LinearRegression(gd)
gd_errors = linear_regression_gd.fit(X, y);
```

```python
y_gd = linear_regression_gd.predict(X)

print("Weights:", linear_regression_gd.get_weights())
print("MSE:", mse(y, y_gd) )

plt.plot(X, y, label='observations')
plt.plot(X, y_gd, label='predictions')
plt.legend()
plt.title('Linear Regression with Gradient Descent')
plt.show()
```

```python
plt.plot(gd_errors, label="RMSE")
plt.legend()
plt.title("RMSE over epochs")
```

### Q1.c
#### Train a Linear Regression model using the Stochastic Gradient Descent

```python
epochs = 3000
learning_rate = 0.005
regularization = 0.0

sgd = StochasticGradientDescent(epochs, learning_rate, regularization)
linear_regression_sgd = LinearRegression(sgd)
sgd_errors = linear_regression_sgd.fit(X, y);
```

```python
y_sgd = linear_regression_sgd.predict(X)

print("Weights:", linear_regression_sgd.get_weights())
print("MSE:", mse(y, y_sgd) )

plt.plot(X, y, label='observations')
plt.plot(X, y_sgd, label='predictions')
plt.legend()
plt.title('Linear Regression with Stochastic Gradient Descent')
plt.show()

plt.plot(sgd_errors)
plt.show()
```
